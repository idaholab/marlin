```python
import torch
import matplotlib.pyplot as plt
import numpy as np
```

```python
p = list(torch.jit.load('grain_labels_raw_2d.pt').parameters())
len(p)
```

```python
fig, axes = plt.subplots(1, 2, figsize=(10, 4), constrained_layout=True)
for i in [0,1]:
  numpy_array = p[0][i,:,:].numpy()
  im = axes[i].imshow(numpy_array, cmap='gray')
  axes[i].set_title(f"p1[{i}]")
  axes[i].set_xlabel("Width")
  axes[i].set_ylabel("Height")
  fig.colorbar(im, ax=axes[i])
plt.show()
```
